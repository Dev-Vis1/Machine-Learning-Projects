 # K Means Clustering

```python
#import imblearn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#from imblearn import  
from sklearn.cluster import KMeans



#dataset
from sklearn.datasets import make_blobs

X,y_true = make_blobs(n_samples=1000,centers=2, cluster_std=1,random_state=123)
```

```python
from plotnine import ggplot, aes, geom_point,geom_line

df = pd.DataFrame(X, columns=['x1','x2'])
df['y'] = y_true

ggplot(df) + aes(x='x1', y='x2', color='y') + geom_point()
```

```python
kmeans = KMeans(n_clusters=3,random_state=123)
kmeans.fit(X)

df_centers = pd.DataFrame(kmeans.cluster_centers_, columns=['x1','x2'])
df_centers
```

```python
y_kmeans = kmeans.predict(X)
df['y_kmeans'] = y_kmeans

(ggplot(df) + aes(x='x1', y='x2', color='y') + geom_point() + geom_point(data=df_centers, color='black',size=3))
```

```python
#elbow method

inertia= []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k, random_state=123)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    
    
df_inertia = pd.DataFrame({'k':range(1,10),'inertia':inertia})
ggplot(df_inertia) + aes(x='k', y='inertia') + geom_point() + geom_line()
```

```python
from sklearn.cluster import DBSCAN
import random
```

```python
num_points = 4000
x = random.sample(population=set(np.linspace(start=-10,stop=10, num = num_points)), k=num_points)
y = random.sample(population=set(np.linspace(start=-10,stop=10, num = num_points)), k=num_points)

z =[(x**2 + y**2) for x,y in zip(x,y)]

df = pd.DataFrame(list(zip(x,y,z)),columns=['x','y','z'])
df['class'] = [1 if ((i<10)|(80<i<100)) else 0 for i in df['z']]

df = df[df['class']==1]

df =df.drop(['z','class'],axis=1)

#visuualization
##
```

```python
clustering =DBSCAN(eps =3,min_samples=2).fit(df)
df['labels'] = [str(i) for i in clustering.labels_]

##plot
```
